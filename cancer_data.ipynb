```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
# seaborn style
sns.set_style('whitegrid')
# Import More Dependenices 
%matplotlib inline
import matplotlib
from datetime import datetime
# warnings 
import warnings
warnings.filterwarnings("ignore")

# Dependencies and Setup
import pandas as pd
import scipy.stats as st
import time
from datetime import datetime, timedelta, date, time
```

```python
csv_path = "Resources/cancer_EN.csv"
europe_df = pd.read_csv(csv_path, encoding="utf-8", header=None)

europe_df.head()
```

```python
europe_df.columns=("Countries", "Sex", "Year", "Value", "Coordinates")

europe_df
```

```python
europe_1_df = europe_df.drop([0,1])
europe_1_df
```

```python
europe_1_df.reset_index(drop=True, inplace=True)
europe_1_df = europe_1_df.astype({'Value':'float'})
europe_1_df
```

```python
country_group = europe_1_df.groupby(["Countries", "Coordinates"])
value_sum = country_group["Value"].mean() 
print(value_sum)
```

```python
cancer_data_EN = pd.DataFrame(value_sum) 
cancer_data_EN_1 = cancer_data_EN.reset_index()
cancer_data_EN_1.rename(columns={"Countries":"Country/State"}, inplace = True)                             
cancer_data_EN_1
```

```python
usa_df = pd.read_excel(r'Resources/cancer_USA.xlsx')

usa_df.head()
```

```python
usa_df.rename(columns={"State":"Country/State"}, inplace = True)

usa_df.head()
```

```python
combined_cancer_df = pd.merge(usa_df, cancer_data_EN_1, on=["Country/State", "Value", "Coordinates"], how="outer").drop(columns=["Sex"])
combined_cancer_df
```

```python
canada_df = pd.read_csv("Resources/cancer_CA.csv", encoding="utf-8", header=None)

canada_df.head()
```

```python
canada_df.columns=("Year", "Country/State", "Sex", "Value", "Coordinates")
canada_df
```

```python
canada_1_df = canada_df.drop([0,1])
canada_1_df.reset_index(drop=True, inplace=True)
canada_1_df = canada_1_df.astype({'Value':'float'})
canada_1_df
```

```python
canada_group = canada_1_df.groupby(["Country/State", "Coordinates"])
value_sum_ca = canada_group["Value"].mean() 
print(value_sum_ca)
```

```python
cancer_data_CA = pd.DataFrame(value_sum_ca) 
cancer_data_CA_1 = cancer_data_CA.reset_index()                             
cancer_data_CA_1
```

```python
final_data = pd.merge(combined_cancer_df, cancer_data_CA_1, on=["Country/State", "Value", "Coordinates"], how="outer")
final_data
```
